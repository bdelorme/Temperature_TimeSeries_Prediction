## Data Visualization

Change date range for zooming/dezooming in the time series.

```python
date_lim_inf = dt.date(2014, 1, 1)
date_lim_sup = dt.date(2014, 12, 31)
#date_lim_inf = dt.date(2014, 9, 22)
#date_lim_sup = dt.date(2014, 9, 28)

f, axarr = plt.subplots(len(data.columns)-1, sharex=True, figsize=(80, 60))
for i in range(1, len(data.columns)):
    axarr[i-1].plot(data.index, data[data.columns[i]])
    axarr[i-1].set_title(data.columns[i], fontsize=60)
    axarr[i-1].xaxis.set_tick_params(labelsize=50)
    axarr[i-1].set_yticks([])
    axarr[i-1].set_xlim([date_lim_inf, date_lim_sup])
```

The time series look rather noisy at first glance. TEMP has an annual cycle with hotter values on average in the summer and colder values in the winter. However, heating occurs at the beginning of each month followed by an opposite cooling event. This behavior occurs over the course of 2 weeks or so. This observation suggests that 'day of the month' and 'month of the year' are two important features. We observe also a diurnal cycle, suggesting that 'hour of the day' is also important. We doo not observe any difference between week days and week-end.

## Features Engineering

```python
data['hour_of_day'] = data.index.hour
data['day_of_month'] = data.index.day
data['month_of_year'] = data.index.month
data.sample(5)
```

## Simple Case: non-autoregressif RF

```python
from sklearn.ensemble import RandomForestRegressor

# fill na
data.fillna(method='bfill', inplace = True)

# separate dataset into train and test
y = data['TEMP'].iloc[1:]
X = data.drop('TEMP', axis=1).iloc[:-1]
Nlim = int(len(y)*0.8)
X_train = X.iloc[:Nlim]
X_test = X.iloc[Nlim:]
y_train = y.iloc[:Nlim]
y_test = y.iloc[Nlim:]

clf = RandomForestRegressor(random_state=0).fit(X_train, y_train)
print('Train score: {}'.format(clf.score(X_train, y_train)))
print('Test score: {}'.format(clf.score(X_test, y_test)))

```

```python
y_pred_test = clf.predict(X_test)
y_pred_train = clf.predict(X_train)

date_lim_inf = dt.date(2016, 3, 1)
date_lim_sup = dt.date(2016, 3, 31)

plt.figure(figsize=(10, 3))
plt.plot(y_train.index, y_train, label="train")
plt.plot(y_test.index, y_test, label="test")
plt.plot(y_train.index, y_pred_train, '--', label="prediction train")
plt.plot(y_test.index, y_pred_test, '--', label="prediction train")
#plt.legend(loc=(1.01, 0))
plt.xlabel("Date")
plt.ylabel("Temp")
plt.xlim([date_lim_inf, date_lim_sup])
```
